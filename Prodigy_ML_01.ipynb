```python
import pandas as pd
from sklearn.metrics import explained_variance_score, mean_squared_error
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
train_data = pd.read_csv('/content/drive/MyDrive/train.csv')
test_data = pd.read_csv('/content/drive/MyDrive/test.csv')
test_labels = pd.read_csv('/content/drive/MyDrive/sample_submission.csv')
```

```python
X_train = train_data[['GrLivArea','BedroomAbvGr','FullBath','HalfBath']]
y_train = train_data['SalePrice']
X_test = test_data[['GrLivArea','BedroomAbvGr','FullBath','HalfBath']]
y_test = test_labels['SalePrice']
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
X_test.dropna(inplace=True)

y_test = y_test[X_test.index]

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
plt.scatter(y_test,y_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices vs Predicted Prices")
plt.show()
```
